```python
import pandas as pd
import numpy as np
import os

parse_data=["FILING_DATE"]
data=pd.read_csv("D://stock_data/all_mapping_0304.csv",parse_dates=parse_data,encoding="ISO-8859-1")
data
```

```python
grouped=data.groupby("DETAIL_JSONcompany_name")

diff1=grouped["avg_sent"].diff().fillna(0)
diff2=grouped["word_count"].diff().fillna(1)
diff3=grouped["sum_sent"].diff().fillna(1)
diff4=grouped["hit_count"].diff().fillna(1)

data['avg_sent_diff']=diff1
data['word_count_diff']=diff2
data['sum_sent_diff']=diff3
data['hit_count_diff']=diff4


# data
```

```python
data["avg_sent_trend"]=data["avg_sent_diff"]/data["word_count_diff"]
data["sum_sent_trend"]=data['sum_sent_diff']/data["word_count_diff"]
data["hit_count_trend"]=data['hit_count_diff']/data["word_count_diff"]
data
```

```python
data["sum_sent_trend"].isna().sum()
```

```python
temp=data
temp.to_csv("D://stock_data/all_factors.csv",header=True)
```
